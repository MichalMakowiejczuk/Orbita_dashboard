```python
from scripts.gpx_parser import GPXParser
from scripts.elevation_profile import ElevationProfile
import matplotlib.pyplot as plt

```

```python
gpx_file = "data/track/orbita25.gpx"
parser = GPXParser(gpx_file)
track_df = parser.parse_to_dataframe()

profile = ElevationProfile(track_df)
profile.geolocate_places(min_distance_km=10, cache_file="cache/places_cache.json")

fig, ax = profile.plot()
fig.set_size_inches(12, 4)
# ax.set_title("Profil wysokości - Orbita 25")
ax.set_xlabel("Dystans [km]")
ax.set_ylabel("Wysokość [m]")
ax.set_ylim(200, profile.track_df["elevation"].max() * 1.1)

plt.show()
```

```python
fig.savefig("static/elevation_profile.png", bbox_inches="tight", dpi=300)
```
